**HISTOGRAM**

```python
#A histogram in Matplotlib is used to represent the distribution of numerical data by dividing the data into bins(intervals) and plotting the frequency of data points in each bin(interval).
#They help in understanding data distribution, detecting outliers, comparing distributions between different groups, and checking assumptions like normality. 
#By adjusting the number of bins, you can control the granularity of the visualization, making it a versatile tool for exploratory data analysis.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ages = np.random.normal(20,1.2,500) #here random.normal() is normal distribution giving random array of sample mean has to be = 20 ,SD =1.2 and total datapoints(people) are 500) 
print(f"ages are:  {ages}")
plt.xlabel("<---------Age----------->")
plt.ylabel("<---------People--------->")
plt.hist(ages ,bins=11 ,color="orange") #graph shows it is normally distributed having mean of 20 
plt.show()
```

```python
plt.xlabel("<---------Age----------->")
plt.ylabel("<---------People--------->")
plt.hist(ages,bins=20,cumulative=True) #it means 500 people are of 24 or below  .Like for 18 ages or below we have approx of 30/40 people if we see histogram graph on x-axis of 18 age
```

```python
#one more example of histogram showing outliers
scores = np.random.normal(70, 10, 1000)
scores = np.append(scores, [20, 100])  # Adding outliers 20 and 100
print(scores)
plt.hist(scores, bins=20, color='purple', edgecolor='black')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Exam Scores Distribution with Outliers')
plt.show()
```

```python
'''
from the above histogram ,we could observe below things :

How to Identify Outliers on the Histogram:

    Main Data Cluster:
        The majority of the scores will be centered around the mean of 70 with a standard deviation of 10, meaning most of the data will fall between 60 and 80. This is where youll see the tallest bars in the histogram.

    Outliers:
        The outliers you added to the data are 20 and 100 only. Since these values are far from the mean (70) and outside the typical range of scores (which would be roughly between 40 and 100), they will appear as very short bars on the far left (around 20) and the far right (around 100) of the histogram.
        Specifically, you should see two small bars, one near the 20 mark and another near the 100 mark. These bars represent the outlier values.

Visual Interpretation:

    The main cluster of the data (most scores) will form a bell-shaped curve centered around 70.
    The bars corresponding to the outliers (20 and 100) will be much lower and positioned far to the left and right of the main cluster, indicating that these values are much less frequent and fall outside the typical range of scores.

This visual separation of the outliers from the main data cluster on the histogram is how you identify them.

'''
```

**BOX PLOT**

```python
'''
A box plot, also known as a box-and-whisker plot, is a graphical representation used to summarize the distribution of a dataset. It provides a visual snapshot of the datas spread, central tendency, and potential outliers. Here's what a box plot typically shows:
Components of a Box Plot:

    Minimum: The smallest data point, excluding outliers.
    First Quartile (Q1): The median of the lower half of the dataset (25th percentile).
    Median (Q2): The middle value of the dataset (50th percentile).
    Third Quartile (Q3): The median of the upper half of the dataset (75th percentile).
    Maximum: The largest data point, excluding outliers.
    Interquartile Range (IQR): The range between the first and third quartiles (Q3 - Q1).
    Whiskers: Lines extending from the box to the minimum and maximum values within 1.5 * IQR from Q1 and Q3, respectively.
    Outliers: Data points that fall outside 1.5 * IQR from Q1 and Q3 are plotted as individual points





Components of a Box Plot:

    Minimum: The smallest data point, excluding outliers.
    First Quartile (Q1): The median of the lower half of the dataset (25th percentile).
    Median (Q2): The middle value of the dataset (50th percentile).
    Third Quartile (Q3): The median of the upper half of the dataset (75th percentile).
    Maximum: The largest data point, excluding outliers.
    Interquartile Range (IQR): The range between the first and third quartiles (Q3 - Q1).
    Whiskers: Lines extending from the box to the minimum and maximum values within 1.5 * IQR from Q1 and Q3, respectively.
    Outliers: Data points that fall outside 1.5 * IQR from Q1 and Q3 are plotted as individual points.

Interpretation:

    Center: The median line inside the box represents the central tendency of the data.
    Spread: The length of the box (IQR) shows the spread of the middle 50% of the data.
    Symmetry: The symmetry of the box and whiskers indicates whether the data is skewed. If the median is closer to Q1, the data is skewed right; if closer to Q3, it's skewed left.
    Outliers: Points outside the whiskers are potential outliers and indicate unusual observations.


Use Case:

    Exploratory Data Analysis (EDA): In EDA, box plots are frequently used to understand the distribution of data before applying more complex analyses.
    Quality Control: In manufacturing or quality control processes, box plots help in identifying variations or defects by showing data spread and outliers.
    Comparative Analysis: When comparing the performance of different models, products, or groups, box plots can provide quick insights into their performance metrics.

Box plots are a powerful tool for summarizing data and spotting key patterns, particularly in comparing multiple distributions.
''' 
```

```python
import numpy as np
import matplotlib.pyplot as plt

heights = np.random.normal(170 , 8 , 500) #mean is 172 (like most of ages is around 172, SD is 8(+- 8 units from 172 mean) and total people 500

plt.boxplot(heights)
plt.show()
```

**SUBPLOTS**

```python
#showing the subplots together
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,101) #it generates 100 integres from 0 to 99
print(x)
print(np.log(x))
fig,axs = plt.subplots(2,2) #subplotting happening here

axs[0,0].plot(x ,np.sin(x))
axs[0,1].plot(x ,np.cos(x))
axs[1,0].plot(x ,np.random.random(100))
axs[1,1].plot(x ,np.log(x))

#now setting title for each subplot needs to be given
axs[0,0].set_title("Sine wave")
axs[0,1].set_title("Cosine wave")
axs[1,0].set_title("Random Function")
axs[1,1].set_title("Log FUnction")

#setting labels
axs[0,0].set_xlabel("TESTING")

#setting main title of these plots
fig.suptitle("Four Plots")

plt.tight_layout()  #by giving this it doesnot overlap the plots and its label
```

**SURFACE 3D PLOTTING**

```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.axes(projection = "3d")

x=np.arange(-5,5,0.1)
y=np.arange(-5,5,.1)

X,Y= np.meshgrid(x,y) #meshgrid is used to create a grid of x and y coordinates for 2D and 3D plotting.
#By using meshgrid, you can easily handle and visualize functions that depend on two variables, making it a powerful tool for scientific computing and data visualization.

Z= np.sin(X) * np.cos(Y)

ax.plot_surface(X,Y,Z, cmap="Spectral") #using plot_surface()
ax.set_title("SURFACE 3D PLOT" ,c="g")
```
