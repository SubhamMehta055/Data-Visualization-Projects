**PAIR PLOT**

```python
'''
A pairplot in Seaborn is a powerful visualization tool used to explore relationships between pairs of variables in a dataset. It creates a grid of scatter plots,
where each plot shows the relationship between two variables, and the diagonal plots often show the distribution of each variable (typically as a histogram or a kernel density plot).

What Pairplot Does:
Visualizes Relationships: Displays pairwise relationships between variables in a dataset.
Distribution Analysis: Shows the distribution of individual variables along the diagonal.
Correlation Insight: Helps in identifying potential correlations between variables.
Categorical Differentiation: Can color-code different categories to analyze how they interact with numerical variables.
'''
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
print(iris)
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Create a pairplot
sns.pairplot(iris_df, hue='species')
plt.show()
```

```python
#In this case, the pairplot will help visualize how each species of flower can be distinguished based on its measurements, aiding in tasks like classification and clustering.
```

**HEATMAP**

```python
'''
A heatmap is a data visualization technique that uses a color-coded grid to represent the values of a dataset. Each cell in the grid corresponds 
to a data point, with the color intensity representing the magnitude of the value in that cell.
Heatmaps are particularly useful for visualizing relationships between variables and identifying patterns, correlations, or areas of high and low intensity.
'''
```

```python
#Heatmap is very useful to spot those features that are very correlated with each other and called collinearity and useful for supervised learning not unsupervised learning
#The higher the correlation and the higher the probability that one or two features that are correlated  with each other may be removed beacuse it may not increase the predictive power of the dataset
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    'GDP': [3.0, 2.5, 4.2, 5.1, 3.8],
    'Inflation': [1.2, 1.3, 1.4, 1.1, 1.5],
    'Unemployment': [7.1, 6.8, 6.5, 5.9, 6.3],
    'Interest Rate': [2.5, 2.7, 2.9, 3.0, 2.6]
}

df = pd.DataFrame(data)
print(df)

# Calculate the correlation matrix
corr_matrix = df.corr() #df.corr(): This function calculates the Pearson correlation coefficient for all pairs of columns in the DataFrame df. and works with linear correlations
#we have one more method called Spearman linear correlation coeff and works with monotonic correlations

'''
To calculate the correlation coefficient between two variables 
𝑋 and 
𝑌, you can use the Pearson correlation formula. The correlation matrix is built by applying this formula to every pair of variables in your dataset.
'''

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')  #annot shows data values in cell

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
'''
Explanation:
Correlation Matrix: The correlation matrix shows the correlation coefficients between pairs of variables, 
ranging from -1 (perfect negative correlation) to +1 (perfect positive correlation). A value of 0 indicates no correlation.

Heatmap: The sns.heatmap() function visualizes this matrix. The annot=True argument adds the correlation values on the heatmap cells, and cmap='coolwarm' specifies the color scheme.
In this example, the heatmap helps to visually identify which variables are strongly correlated (either positively or negatively) and which are not. 
This insight can guide further analysis or feature selection in machine learning models.
'''
```

```python
'''
In Above heatmap diagram interpretations are:

A correlation of 0.85 between GDP and Inflation suggests a strong positive linear relationship.
A correlation of -0.92 between GDP and Unemployment suggests a strong negative linear relationship.
A correlation of 0.75 between GDP and Interest Rate suggests a moderate positive linear relationship.

'''
```

**CLUSTER MAP**

```python
'''clustermaps are particularly useful in fields like bioinformatics, where understanding the
relationships between multiple variables (e.g., gene expressions) is crucial for gaining insights into biological processes.
'''
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Sample gene expression data
data = {
    'Gene1': [1.2, 0.8, 2.4, 1.5, 2.1],
    'Gene2': [3.2, 3.3, 2.9, 3.5, 3.1],
    'Gene3': [0.5, 0.6, 0.4, 0.7, 0.8],
    'Gene4': [1.9, 2.0, 2.1, 1.8, 1.7],
    'Gene5': [2.4, 2.2, 2.5, 2.3, 2.1],
}

# Convert to DataFrame
df = pd.DataFrame(data, index=['Sample1', 'Sample2', 'Sample3', 'Sample4', 'Sample5'])

# Create a clustermap
sns.clustermap(df, standard_scale=1)

plt.show()
```
